```julia
using JLD
using DataFrames
using DataFramesMeta
using Plots
using LaTeXStrings
pgfplots()
default(size=(600,400))
```

```julia
function analyze_error_grad()

    df_primal_ok = load("dataframes/primal_l_study.jld")["df"]
    df_dual_ok = load("dataframes/dual_l_study.jld")["df"]
    df_dual = load("dataframes/dual_data_frame.jld")["df"]
    df_primal = load("dataframes/primal_data_frame.jld")["df"]

    p1 = plot()
    p2 = plot()
    for (i, l) in enumerate(0.1:0.1:0.5)
        df_dual_l = @where(df_dual, :l .== l)
        df_primal_l = @where(df_primal, :l .== l)
        df_primal_ok_l = @where(df_primal_ok, :l.==l)
        df_dual_ok_l = @where(df_dual_ok, :l.==l)

        eles = df_dual_l[:n_elements]
        primal_err_grad_relative = df_primal_l[:err_grad_energy].^2 ./  df_primal_ok_l[:tot_grad_energy].^2
        dual_err_grad_relative = df_dual_l[:err_grad_energy].^2 ./  df_dual_ok_l[:tot_grad_energy].^2

        for (p, name, var) in  ((p1, "Dual", dual_err_grad_relative),
                                (p2, "Primal", primal_err_grad_relative))
        
            plot!(p, eles, var, 
            label=LaTeXStrings.latexstring("l = $l"),
            xlabel ="number of cells",
            ylabel =L"\Psi_g(\mathbf{\xi} - \mathbf{\xi}_h) / \Psi_g(\mathbf{\xi})",
            title = "$name: Relative error in gradient energy",
            yscale=:log
            )
        end
        

    end
    
 
    
    return p1, p2
end
```

```julia
p1, p2 = analyze_error_grad();
```

```julia
function analyze_error_slip()

    df_primal_ok = load("dataframes/primal_l_study.jld")["df"]
    df_dual_ok = load("dataframes/dual_l_study.jld")["df"]
    df_dual = load("dataframes/dual_data_frame.jld")["df"]
    df_primal = load("dataframes/primal_data_frame.jld")["df"]

    p = plot()
    p2 = plot()
    for (i, l) in enumerate(unique(df_dual[:l]))
        df_dual_l = @where(df_dual, :l .== l)
        df_primal_l = @where(df_primal, :l .== l)
        df_primal_ok_l = @where(df_primal_ok, :l.==l)
        df_dual_ok_l = @where(df_dual_ok, :l.==l)

        eles = df_dual_l[:n_elements]
        primal_err_grad_relative = df_primal_l[:err_slip] ./  df_primal_ok_l[:tot_slip]
        dual_err_grad_relative = df_dual_l[:err_slip] ./  df_dual_ok_l[:tot_slip]

      #  plot!(p, eles, dual_err_grad_relative, label ="$l")
        plot!(p2, eles, primal_err_grad_relative, 
        label=LaTeXStrings.latexstring("l = $l"),
        xlabel ="number of cells",
        ylabel =L" \sqrt{\sum [\gamma_\alpha - \gamma_{\alpha, h}]^2} / \sqrt{\sum [\gamma_\alpha]^2}",
        title = "Primal: Error in effective slip"
        )
        
        #PyPlot.savefig("/home/kristoffer/Dropbox/PhD/Research/Papers/Paper_I/4734916dmgkzg/figs/primal_slip_err.pdf")

        
        plot!(p, eles, dual_err_grad_relative, 
        label=LaTeXStrings.latexstring("l = $l"),
        xlabel ="number of cells",
        ylabel =L" \sqrt{\sum [\gamma_\alpha - \gamma_{\alpha, h}]^2} / \sqrt{\sum [\gamma_\alpha]^2}",
        title = "Dual: Error in effective slip"
        )
        
        #PyPlot.savefig("/home/kristoffer/Dropbox/PhD/Research/Papers/Paper_I/4734916dmgkzg/figs/dual_slip_err.pdf")
    end
    return p2, p
end
```

```julia
pgfplots()
 function analyze_grad_energy_with_l()
    df = load("dataframes/primal_l_study.jld")["df"]
    ps = plot(df[:l], df[:tot_slip],
            ylabel = L"$\overline{\gamma}$",
            xlabel = L"$l$",
            title = L"Effective slip $\overline{\gamma} = \sqrt{\int \sum_{\alpha} \gamma_{\alpha}^2 dV}$",
            legend = false
    )
    pg = plot(df[:l], df[:tot_grad_energy],
        ylabel = L"$\Psi_g$",
        xlabel = L"$l$",
        title = "Gradient energy",
        legend = false
    )
    return ps, pg
end

ps, pg = analyze_grad_energy_with_l()
```
